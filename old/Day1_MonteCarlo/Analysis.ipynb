```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('white')
```

```python
df=pd.read_csv("random_matrix.csv")
#print only the first few elements
print(df)
```

```python
g=sns.FacetGrid(df,col='n')
g.map(sns.distplot,'eigenvalues')
g.set_axis_labels("Eigenvalue","Probability density")
plt.savefig("eigenvalue_distribution.pdf",bbox_inches='tight')
```
